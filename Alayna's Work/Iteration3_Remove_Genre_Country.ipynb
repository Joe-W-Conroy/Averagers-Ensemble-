```python
import numpy as np
import pandas as pd

movies = pd.read_csv('IMDb_movies.csv', dtype={'country': str, 'genre': str})
print(movies.shape)
movies.head()
```

```python
movies.describe()
```

```python
movies.dropna(subset=['genre','country','avg_vote','budget','usa_gross_income','worlwide_gross_income','metascore'], inplace=True)
movies.drop_duplicates(['title'], inplace=True)
movies = movies.drop(['title','original_title','date_published','duration','language','production_company','description','reviews_from_users','reviews_from_critics'], axis=1)
movies = movies.drop(['director','writer','actors'], axis=1)
```

```python
print(movies.shape)
movies.head()
```

```python
movies['year'] = movies['year'].astype(int)
movies['avg_vote'] = movies['avg_vote'].astype(float)
movies['votes'] = movies['votes'].astype(int)
```

```python
movies = movies[movies['year'] >= 1970]
movies.head()
```

```python
import re

movies.loc[:,'genre'] = movies.loc[:,'genre'].apply(lambda x: re.sub('[^a-zA-z0-9\s]','',x))
movies.loc[:,'country'] = movies.loc[:,'country'].apply(lambda x: re.sub('[^a-zA-z0-9\s]','',x))

movies.head()
```

```python
def lower_case(input_str):
    input_str = input_str.lower()
    return input_str

movies.loc[:,'genre'] = movies.loc[:,'genre'].apply(lambda x: lower_case(x))
movies.loc[:,'country'] = movies.loc[:,'country'].apply(lambda x: lower_case(x))
movies.head()
```

```python
movies = movies.drop(['genre','country'], axis=1)
```

```python
usd = movies[movies['budget'].str.contains("$")]
usd.loc[:,'budget'] = usd.loc[:,'budget'].apply(lambda x: re.sub('[^0-9\s]','',x))
usd.loc[:,'usa_gross_income'] = usd.loc[:,'usa_gross_income'].apply(lambda x: re.sub('[^0-9\s]','',x))
usd.loc[:,'worlwide_gross_income'] = usd.loc[:,'worlwide_gross_income'].apply(lambda x: re.sub('[^0-9\s]','',x))
print(usd.shape)
usd.head()
```

```python
usd['budget'] = usd['budget'].astype(float)
usd['usa_gross_income'] = usd['usa_gross_income'].astype(float)
usd['worlwide_gross_income'] = usd['worlwide_gross_income'].astype(float)
usd.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
import pydotplus #pip install pydotplus
from sklearn.tree import export_graphviz
from sklearn.model_selection import train_test_split

usd_train, usd_test = train_test_split(usd, test_size=0.2, random_state=42)

print("usd_train size:", len(usd_train))
print("usd_test size:", len(usd_test), "\n")
```

```python
usd_labels1 = usd_train['avg_vote']
usd_labels2 = usd_test['avg_vote']
print(usd_labels1.shape)
print(usd_labels2.shape)
```

```python
usd_train = usd_train.drop(['avg_vote','imdb_title_id'], axis=1)
usd_test = usd_test.drop(['avg_vote','imdb_title_id'], axis=1)
usd_train.head()
```

```python
usd_train.dtypes
```

```python
usd_train = usd_train.reset_index(drop = True)
usd_test = usd_test.reset_index(drop = True)
usd_labels1 = usd_labels1.reset_index(drop = True)
usd_labels2 = usd_labels2.reset_index(drop = True)
```

```python
clf_tree = DecisionTreeRegressor()
clf_tree.fit(usd_train, usd_labels1)
pred = clf_tree.predict(usd_test)
print(pred)
print(usd_labels2)
```

```python
conf = np.zeros(1300)
y = usd_labels2 - pred
y = abs(y)
y = y.reset_index(drop = True)

for i in range(1300):
    if y[i]<0.5 or y[i]==0.5:
        conf[i] = 1
        
print(y)
print(conf)
test = np.ones(1300)
print(test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('Accuracy score: {}'.format(accuracy_score(conf,test)))
print('F1 score: {}'.format(f1_score(conf,test)))
```
